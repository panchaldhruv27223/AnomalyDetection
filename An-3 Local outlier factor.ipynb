```python
import numpy as np 
import pandas as pd
from sklearn.datasets import make_circles
```

```python
X,y = make_circles(n_samples=1000,factor=0.3,noise=0.15)
X,y
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X[:,0],X[:,1])
```

```python
X
```

```python
from sklearn.neighbors import LocalOutlierFactor
```

```python
lof = LocalOutlierFactor(n_neighbors=7)
```

```python
out = lof.fit_predict(X)
```

```python
out
```

```python
plt.scatter(X[:,0],X[:,1],c=out)
```
